```python
# normalizing text - punctuation, capitalization, lemmatization

nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def normalize_text(text):
    lowered = text.lower()
    nopunc = re.sub(r'[^\w\s]', '', lowered)
    tokens = word_tokenize(nopunc)
    words = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(words)

titles_normalized = [normalize_text(title) for title in claude_titles_2025]
pairs_normalized = [normalize_text(title) for title in claude_paired_titles]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def similarity_filter(normalized_titles, threshold = 0.15):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(normalized_titles)
    cosine_sim = cosine_similarity(tfidf_matrix)

    to_remove = set()
    for i in range(len(cosine_sim)):
        for j in range(i + 1, len(cosine_sim)):
            if cosine_sim[i, j] > threshold:
                to_remove.add(j)

    filtered= [normalized_titles[i] for i in range(len(normalized_titles)) if i not in to_remove]
    return filtered

similarity_filter(pairs_normalized)
```
