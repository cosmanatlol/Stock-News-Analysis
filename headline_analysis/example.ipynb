```python
%load_ext autoreload
%autoreload 2
import pandas as pd
from llm.llm_models import headline_rating, relevance_extractor 
import time
healthcare = ['ISRG', 'LLY', 'IDXX', 'BSX', 'EW', 'DHR', 'SYK', 'CI', 'BMY', 'CVS']
energy = ['WMB', 'KMI', 'OKE', 'OXY', 'CVX', 'XOM', 'PSX', 'EOG', 'FANG', 'SLB']
technology = ['PLTR', 'CRWD', 'NOW', 'FICO', 'PANW', 'ANET', 'CDNS', 'QCOM', 'DELL', 'MU']
stocks = {'healthcare': healthcare, 'energy': energy, 'technology': technology}
path = 'data.csv'
data = pd.read_csv(path)
data['date'] = pd.to_datetime(data['date']).dt.day
date_range = data['date'].unique()
final_data = []
for days in date_range:
    data_splice = data[data['date'] == days]
    for sector in list(stocks.keys()):
        relevance_list = []
        for x in data_splice['headline']:
            relevance_list += [relevance_extractor(x, sector)]
            time.sleep(1)
        headlines = data_splice[relevance_list]['headline'].tolist()
        for stock in stocks[sector]:
            rating = headline_rating(data_splice['headline'][relevance_list].tolist(), stock)
            final_data.append([stock, days, rating])
            time.sleep(1)
final_df = pd.DataFrame(final_data, columns=['stock', 'date', 'rating'])
final_df
```
